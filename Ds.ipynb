```python
def review(data):
    print('\n' ,20*'*', ' COLUMNS NAMES ', '*'*20, '\n')
    print(data.columns)
    print('\n' ,20*'*', ' INFO ', '*'*20, '\n')
    print(data.info())
    print('\n' ,20*'*', ' UNIQUE VALUES ', '*'*20, '\n')
    print(data.nunique())
    print('\n' ,20*'*', ' MSSING VALUES ', '*'*20, '\n')
    print(data.isna().sum())
    print('\n' ,20*'*', ' DUPLICATED VALUES ', '*'*20, '\n')
    print(data.duplicated())
    print('\n' ,20*'*', ' DESCRIPTIVE STATS DATA ', '*'*20, '\n')
    print(data.describe())
    print('\n' ,20*'*', ' CORRELATION MATRIX ', '*'*20, '\n')
    print(data.corr(numeric_only=True))
```

```python
review(df)
```
